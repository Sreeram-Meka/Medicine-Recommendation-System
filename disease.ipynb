```python
import numpy as np
import pandas as pd
import pickle
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Step 1: Load and Prepare Data
data = pd.read_csv('files/Training.csv')

# Extract features and target variable
X = data.drop(columns=['prognosis'])
y = data['prognosis']

# Encode the target variable
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Step 2: Split Data for Training and Testing
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

# Step 3: Train the Model
svc = SVC(kernel='linear', probability=True)
svc.fit(X_train, y_train)

# Step 4: Evaluate the Model
y_pred = svc.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
print("Classification Report:\n", classification_report(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))

# Step 5: Save Model and Other Dependencies
model_data = {
    "model": svc,
    "label_encoder": label_encoder,
    "symptoms_dict": {symptom: idx for idx, symptom in enumerate(X.columns)},
    "diseases_list": {idx: disease for idx, disease in enumerate(label_encoder.classes_)}
}
pickle.dump(model_data, open('my_models/svc_model_data.pkl', 'wb'))
```
